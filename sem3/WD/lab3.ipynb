```python
# Wizualizacja Danych, Lab 3 - Jan Banot
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, widgets
```

```python
# Zadanie 1
diamonds = sns.load_dataset("diamonds")

plt.figure(figsize=(12, 8))
sns.histplot(data=diamonds, x="price", hue="cut", kde=True, palette="viridis")
plt.title("Rozkład cen diamentów z podziałem na jakość cięcia")
plt.xlabel("Cena")
plt.ylabel("Liczba diamentów")
plt.legend(title="Jakość cięcia")
plt.show()
```

```python
# Zadanie 2
flights = sns.load_dataset("flights")

g = sns.displot(
    data=flights, x="passengers", col="month", col_wrap=4, kde=True, palette="viridis"
)
g.set_axis_labels("Liczba pasażerów", "Częstość")
g.set_titles("Miesiąc: {col_name}")
plt.suptitle("Rozkład liczby pasażerów w różnych miesiącach", y=1.02)

plt.show()
```

```python
# Zadanie 3
titanic = sns.load_dataset("titanic")

plt.figure(figsize=(12, 8))
sns.violinplot(data=titanic, x="sex", y="age", hue="class", palette="muted")
plt.title("Rozkład wieku w zależności od płci i klasy na Titanicu")
plt.xlabel("Płeć")
plt.ylabel("Wiek")
plt.legend(title="Klasa")

plt.show()
```

```python
# Zadanie 4
plt.figure(figsize=(12, 8))
sns.boxplot(
    data=diamonds,
    x="color",
    y="carat",
    palette="coolwarm",
    order=["D", "E", "F", "G", "H", "I", "J"],
)
plt.title("Rozkład masy diamentów (karat) w podziale na jakość koloru")
plt.xlabel("Jakość koloru")
plt.ylabel("Masa (karat)")

plt.show()
```

```python
# Zadanie 6
def plot_trig_function(func_name, x_min, x_max):
    x = np.linspace(x_min, x_max, 500)
    plt.figure(figsize=(10, 6))

    if func_name == "sin":
        y = np.sin(x)
        plt.plot(x, y)
        plt.title("Wykres funkcji sinus")
    elif func_name == "cos":
        y = np.cos(x)
        plt.plot(x, y)
        plt.title("Wykres funkcji cosinus")
    elif func_name == "tan":
        y = np.tan(x)
        y[np.abs(np.cos(x)) < 1e-10] = np.nan
        plt.plot(x, y)
        plt.title("Wykres funkcji tangens")
        plt.ylim(-10, 10)

    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.show()


interact(
    plot_trig_function,
    func_name=widgets.Dropdown(
        options=["sin", "cos", "tan"], value="sin", description="Funkcja:"
    ),
    x_min=widgets.FloatSlider(
        value=-10, min=-20, max=20, step=0.5, description="x min:"
    ),
    x_max=widgets.FloatSlider(
        value=10, min=-20, max=20, step=0.5, description="x max:"
    ),
)
```

```python
# Zadanie 7
def create_sales_chart():
    months = ["Styczeń", "Luty", "Marzec", "Kwiecień", "Maj", "Czerwiec"]
    sales = [45000, 52000, 48000, 61000, 58000, 67000]

    plt.figure(figsize=(10, 6))
    plt.bar(months, sales, color="steelblue")
    plt.title("Sprzedaż w pierwszym półroczu")
    plt.xlabel("Miesiąc")
    plt.ylabel("Sprzedaż (PLN)")
    plt.xticks(rotation=45)
    plt.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.show()


def create_profit_chart():
    months = ["Styczeń", "Luty", "Marzec", "Kwiecień", "Maj", "Czerwiec"]
    profit = [12000, 15000, 13500, 18000, 16500, 21000]

    plt.figure(figsize=(10, 6))
    plt.bar(months, profit, color="seagreen")
    plt.title("Zysk w pierwszym półroczu")
    plt.xlabel("Miesiąc")
    plt.ylabel("Zysk (PLN)")
    plt.xticks(rotation=45)
    plt.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.show()


def create_customers_chart():
    months = ["Styczeń", "Luty", "Marzec", "Kwiecień", "Maj", "Czerwiec"]
    customers = [320, 385, 360, 425, 410, 470]

    plt.figure(figsize=(10, 6))
    plt.bar(months, customers, color="coral")
    plt.title("Liczba klientów w pierwszym półroczu")
    plt.xlabel("Miesiąc")
    plt.ylabel("Liczba klientów")
    plt.xticks(rotation=45)
    plt.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.show()


tab = widgets.Tab()
tab.children = [
    widgets.interactive_output(create_sales_chart, {}),
    widgets.interactive_output(create_profit_chart, {}),
    widgets.interactive_output(create_customers_chart, {}),
]
tab.set_title(0, "Sprzedaż")
tab.set_title(1, "Zysk")
tab.set_title(2, "Liczba klientów")

display(tab)
```

```python
# Zadanie 8
np.random.seed(42)
x_data = np.arange(1, 21)
y_data = np.random.randint(10, 100, size=20)


def plot_chart(chart_type):
    plt.figure(figsize=(10, 6))

    if chart_type == "scatter":
        plt.scatter(x_data, y_data, s=100, c="steelblue", alpha=0.6, edgecolors="black")
        plt.title("Wykres punktowy - dane losowe")
    elif chart_type == "line":
        plt.plot(
            x_data, y_data, marker="o", color="seagreen", linewidth=2, markersize=8
        )
        plt.title("Wykres liniowy - dane losowe")
    elif chart_type == "bar":
        plt.bar(x_data, y_data, color="coral", edgecolor="black")
        plt.title("Wykres słupkowy - dane losowe")

    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


interact(
    plot_chart,
    chart_type=widgets.Dropdown(
        options=["scatter", "line", "bar"], value="scatter", description="Typ wykresu:"
    ),
)
```

```python
# Zadanie 9
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

mpg = sns.load_dataset("mpg")

manufacturers = sorted(mpg["origin"].dropna().unique().tolist())
drive_types = ["All"] + sorted(mpg["origin"].dropna().unique().tolist())


def plot_mpg_scatter(manufacturer, drive):
    plt.figure(figsize=(10, 6))

    filtered_data = mpg.copy()

    if manufacturer != "All":
        filtered_data = filtered_data[filtered_data["origin"] == manufacturer]

    if drive != "All":
        filtered_data = filtered_data[filtered_data["origin"] == drive]

    plt.scatter(
        filtered_data["horsepower"],
        filtered_data["mpg"],
        s=80,
        alpha=0.6,
        c="steelblue",
        edgecolors="black",
    )

    title = "Horsepower vs MPG"
    if manufacturer != "All" or drive != "All":
        filters = []
        if manufacturer != "All":
            filters.append(f"Origin: {manufacturer}")
        if drive != "All":
            filters.append(f"Drive: {drive}")
        title += f" ({', '.join(filters)})"

    plt.title(title, fontsize=14)
    plt.xlabel("Horsepower", fontsize=12)
    plt.ylabel("MPG (Miles per Gallon)", fontsize=12)
    plt.grid(True, alpha=0.3)

    plt.text(
        0.02,
        0.98,
        f"Liczba punktów: {len(filtered_data)}",
        transform=plt.gca().transAxes,
        verticalalignment="top",
        bbox=dict(boxstyle="round", facecolor="wheat", alpha=0.5),
    )

    plt.tight_layout()
    plt.show()


interact(
    plot_mpg_scatter,
    manufacturer=widgets.Dropdown(
        options=["All"] + manufacturers, value="All", description="Origin:"
    ),
    drive=widgets.Dropdown(options=drive_types, value="All", description="Drive:"),
)
```

```python
# Zadanie 10
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

titanic = sns.load_dataset("titanic")

numeric_columns = ["age", "fare", "pclass"]
categorical_columns = ["sex", "class", "who", "embarked", "survived", "pclass", "alone"]


def plot_titanic_data(column, plot_type):
    plt.figure(figsize=(10, 6))

    if plot_type == "histogram":
        if column in numeric_columns:
            plt.hist(
                titanic[column].dropna(),
                bins=30,
                color="steelblue",
                edgecolor="black",
                alpha=0.7,
            )
            plt.title(f"Histogram - {column}", fontsize=14)
            plt.xlabel(column, fontsize=12)
            plt.ylabel("Częstość", fontsize=12)
        else:
            plt.text(
                0.5,
                0.5,
                f'Histogram nie jest odpowiedni dla kolumny "{column}".\nWybierz "countplot" lub "barplot".',
                ha="center",
                va="center",
                fontsize=12,
                transform=plt.gca().transAxes,
            )
            plt.title(f"Błąd wizualizacji - {column}", fontsize=14)

    elif plot_type == "barplot":
        if column in categorical_columns:
            sns.barplot(
                data=titanic, x=column, y="age", errorbar=None, palette="viridis"
            )
            plt.title(f"Barplot - średni wiek według {column}", fontsize=14)
            plt.xlabel(column, fontsize=12)
            plt.ylabel("Średni wiek", fontsize=12)
            plt.xticks(rotation=45)
        else:
            # Dla kolumn numerycznych pokazujemy rozkład
            data_sorted = titanic[column].value_counts().sort_index()
            plt.bar(
                data_sorted.index,
                data_sorted.values,
                color="seagreen",
                edgecolor="black",
            )
            plt.title(f"Barplot - rozkład {column}", fontsize=14)
            plt.xlabel(column, fontsize=12)
            plt.ylabel("Liczba wystąpień", fontsize=12)

    elif plot_type == "countplot":
        sns.countplot(data=titanic, x=column, palette="muted")
        plt.title(f"Countplot - liczba wystąpień {column}", fontsize=14)
        plt.xlabel(column, fontsize=12)
        plt.ylabel("Liczba", fontsize=12)
        plt.xticks(rotation=45)

    plt.grid(True, alpha=0.3, axis="y")
    plt.tight_layout()
    plt.show()


all_columns = numeric_columns + [
    col for col in categorical_columns if col not in numeric_columns
]

interact(
    plot_titanic_data,
    column=widgets.Dropdown(
        options=sorted(all_columns), value="age", description="Kolumna:"
    ),
    plot_type=widgets.Dropdown(
        options=["histogram", "barplot", "countplot"],
        value="histogram",
        description="Typ wykresu:",
    ),
)
```
